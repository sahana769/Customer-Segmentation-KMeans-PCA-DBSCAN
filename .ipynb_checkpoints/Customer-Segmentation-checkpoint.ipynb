# “Customer Segmentation using KMeans, PCA, and DBSCAN”

## Problem Statement

Businesses collect large volumes of customer transaction data, but without proper analysis, it is difficult to understand customer behavior and preferences.

The goal of this project is to perform **unsupervised customer segmentation** by grouping customers with similar purchasing patterns. This project uses multiple advanced clustering techniques:

- **KMeans Clustering** to identify distinct customer segments  
- **PCA (Principal Component Analysis)** for dimensionality reduction and feature optimization  
- **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)** to detect anomalies or unusual customer behavior  

The insights gained from this analysis can help businesses make data-driven decisions, including:

- Targeted marketing campaigns  
- Personalized promotions  
- Customer retention strategies  
- Revenue optimization

By combining clustering and dimensionality reduction, this project provides a comprehensive understanding of customer segments and identifies high-value or anomalous customers in the dataset.


### Step 1: Import libraries

```python
# Data manipulation
import pandas as pd
import numpy as np

# Data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Preprocessing / Scaling
from sklearn.preprocessing import StandardScaler

# Dimensionality Reduction
from sklearn.decomposition import PCA

# Clustering algorithms
from sklearn.cluster import KMeans, DBSCAN

# Clustering evaluation
from sklearn.metrics import silhouette_score

# Ignore warnings for cleaner output
import warnings
warnings.filterwarnings('ignore')

# Optional: display plots inline in Jupyter
%matplotlib inline
```

### Step 2: Load the Dataset

```python
# Load dataset
df = pd.read_csv('UK E-commerce.csv', encoding='ISO-8859-1')

# View first 5 rows
df.head()
```

```python
# Check data types and missing values
df.info()
```

```python
# Summary statistics for numeric columns
df.describe()
```

```python
# Count missing values
df.isnull().sum()
```

```python
df.duplicated().sum()
```

### Step 3: Initial Data Cleaning

```python
# Remove rows without CustomerID
df = df.dropna(subset=['CustomerID'])
```

```python
#Remove cancelled transactions
#In this dataset, cancelled invoices start with 'C' in InvoiceNo
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

### Step 4: RFM Feature Engineering

```python
# Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.info()
```

```python
# Set a reference date for Recency
# Usually the day after the last transaction in dataset
import datetime as dt

# Reference date = max date + 1 day
reference_date = df['InvoiceDate'].max() + dt.timedelta(days=1)
reference_date
```

```python
# Create RFM metrics per customer
# Aggregate data per CustomerID
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (reference_date - x.max()).days,  # Recency
    'InvoiceNo': 'nunique',  # Frequency
    'Quantity': 'sum',  # Optional check, can use Monetary below
    'UnitPrice': 'sum'    # We'll calculate Monetary as total spent
})

# Calculate Monetary properly
rfm['Monetary'] = df.groupby('CustomerID').apply(lambda x: (x['Quantity'] * x['UnitPrice']).sum())

# Rename columns
rfm.rename(columns={'InvoiceDate': 'Recency', 'InvoiceNo': 'Frequency'}, inplace=True)

# Keep only relevant columns
rfm = rfm[['Recency', 'Frequency', 'Monetary']]

rfm.head()
```

```python
# Check distributions
# Visualize distributions
plt.figure(figsize=(15,4))
plt.subplot(1,3,1)
sns.histplot(rfm['Recency'], bins=30)
plt.title('Recency')

plt.subplot(1,3,2)
sns.histplot(rfm['Frequency'], bins=30)
plt.title('Frequency')

plt.subplot(1,3,3)
sns.histplot(rfm['Monetary'], bins=30)
plt.title('Monetary')

plt.show()
```

### Step 5: Feature Scaling

```python
# Since clustering is distance-based, scale all features:
# StandardScaler
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm)

# Convert back to DataFrame
rfm_scaled = pd.DataFrame(rfm_scaled, index=rfm.index, columns=rfm.columns)
rfm_scaled.head()
```

### Step 6: Dimensionality Reduction using PCA

```python
# Apply PCA to scale features for visualization
from sklearn.decomposition import PCA

# Reduce to 2 components for 2D visualization
pca = PCA(n_components=2)
rfm_pca = pca.fit_transform(rfm_scaled)

# Convert to DataFrame
rfm_pca_df = pd.DataFrame(rfm_pca, columns=['PCA1', 'PCA2'], index=rfm.index)
rfm_pca_df.head()
```

```python
# Plot PCA components
plt.figure(figsize=(8,6))
sns.scatterplot(x='PCA1', y='PCA2', data=rfm_pca_df)
plt.title('PCA of Customers')
plt.show()
```

### Step 7: KMeans Clustering

```python
# Find the optimal number of clusters using Elbow Method
from sklearn.cluster import KMeans

# Try k values from 2 to 10
sse = []
for k in range(2, 11):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(rfm_scaled)
    sse.append(km.inertia_)  # Sum of squared distances

# Plot the Elbow Curve
plt.figure(figsize=(8,5))
plt.plot(range(2,11), sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
# Apply KMeans with the chosen k
# Suppose optimal k = 4
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(rfm_scaled)

# Assign cluster labels
rfm['KMeans_Cluster'] = kmeans.labels_
rfm.head()
```

```python
# Evaluate cluster quality using Silhouette Score
from sklearn.metrics import silhouette_score

score = silhouette_score(rfm_scaled, rfm['KMeans_Cluster'])
print(f'Silhouette Score for KMeans: {score:.3f}')
```

```python
# Visualize KMeans clusters (after PCA)
# Add cluster labels to PCA dataframe
rfm_pca_df['Cluster'] = rfm['KMeans_Cluster']

plt.figure(figsize=(8,6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', palette='Set2', data=rfm_pca_df)
plt.title('Customer Segments (KMeans)')
plt.show()
```

### Step 8: Anomaly Detection with DBSCAN

```python
# Apply DBSCAN on scaled RFM features
# Initialize DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=5)  # eps may need adjustment based on your data
dbscan_labels = dbscan.fit_predict(rfm_scaled)

# Add labels to RFM dataframe
rfm['DBSCAN_Cluster'] = dbscan_labels

# Count of each cluster (-1 is considered as outlier)
rfm['DBSCAN_Cluster'].value_counts()
```

```python
# Visualize DBSCAN clusters (after PCA)
# Add DBSCAN labels to PCA dataframe
rfm_pca_df['DBSCAN_Cluster'] = rfm['DBSCAN_Cluster']

plt.figure(figsize=(8,6))
sns.scatterplot(x='PCA1', y='PCA2', hue='DBSCAN_Cluster', palette='Set1', data=rfm_pca_df)
plt.title('Customer Segments + Anomalies (DBSCAN)')
plt.show()
```
